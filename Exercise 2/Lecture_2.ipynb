# Example 5: Booleans (True/False)

**Booleans** represent truth values and are the backbone of decisions in code.

- Two values only: `True` and `False` (note the capital T/F).
- Created by comparisons: `3 < 5` â†’ `True`, `age >= 18` â†’ `True/False`.
- They can be stored in variables: `is_adult = age >= 18`.

**Truthiness in Python**

Some values behave like False in conditions: `0`, `0.0`, `""`, `[]`, `{}`, `None`. Everything else is treated as True.
> For clarity , prefer explicit comparisons: `len(name) > 0` rather than just `if name:`.


````python
age_anna = 10
age_tom = 5
is_tom_younger_than_anna = age_tom < age_anna
print(is_tom_younger_than_anna) # What is the result?
````

````python
is_tom_older_than_anna = age_tom > age_anna
print(is_tom_older_than_anna) # What is the result?
````

# Example 9: Control structures


## If/else
**If/else = decision making.**

- A condition like `my_age < 18` evaluates to `True` or `False` (type `bool`).
- Indentation defines the *block* of code that runs when the condition is true.
- Use `if/elif/else` to cover multiple branches.

````python
my_age = 18

print("Me: A schnaps, please!")

if my_age < 18: # what is the data type of the condition?
    print("Waiter: Sorry, you're too young kid!")
else:
    print("Waiter: Sure, I'll bring you one.")
````

### Check the data type of the condition 

> Tip: print the condition value first to see why a branch runs: `print(my_age < 18)`.

````python
my_age = 18

print("Me: A schnaps, please!")

under_eighteen = my_age < 18
print(under_eighteen)

if under_eighteen:
    print("Waiter: Sorry, you're too young kid!")
else:
    print("Waiter: Sure, I'll bring you one.")
````

## Common intendation errors

**Indentation = structure.**

Python uses **indentation (spaces at the start of a line)** to mark code blocks. Lines with the same indentation level belong together.

- Standard: 4 spaces per level or tab.
- After `if`, `for`, `while`, `def`, add a **colon** `:` and indent the next lines.
- All lines at one level must align exactly.

Common issues:
- Mixing tabs and spaces causes errors â†’ set your editor to insert spaces.
- Misaligned blocks â†’ re-indent the whole block consistently.

> If you get `IndentationError`, compare spaces visually or re-indent the block.

````python
my_age = 18

print("Me: A schnaps, please!")

if my_age >= 18:
    print("Waiter: Sure, I'll bring you one.") # missing intendation
else:
    print("Waiter: Sorry, you're too young kid!")
    print("Waiter: Can I offer you a beer instead?") 
````

## Nested if statements

**Nesting** lets you check conditions inside conditions.

- **Only the first condition which is `True` gets executed**

> Use sparingly; deep nesting is hard to read. Prefer `elif` or combine conditions when possible.

### Waiter offering an alternative, if we are too young to drink schnaps, but allowed to drink beer.

````python
my_age = 15

print("Me: A schnaps, please!")

is_eighteen = my_age >= 18
is_sixteen = my_age >= 16

print(is_eighteen)
print(is_sixteen)

if is_sixteen:
    if is_eighteen:
        print("Waiter: Sure, I'll bring you one.")
    else:
        print("Waiter: Sorry, you're too young to drink schnaps, but can I offer you a beer?")
else:
    print("Waiter: Sorry, you're too young kid!")
````

### There are multiple ways to solve the same problem

````python
my_age = 17
print("Me: A schnaps, please!")

is_eighteen = my_age >= 18
is_sixteen = my_age >= 16

print(is_eighteen)
print(is_sixteen)

if is_eighteen: # only the FIRST condition which is TRUE gets executed
    print("Waiter: Sure, I'll bring you one.")
elif is_sixteen:
    print("Waiter: Sorry, you're too young to drink schnaps, but can I offer you a beer?")
else:
    print("Waiter: Sorry, you're too young kid!")
````

#### Bonus example for motivated students

You can combine several conditions using **logical operators** like `and`, `or`, and `not`.

````python
my_age = 18
print("Me: A schnaps, please!")

is_eighteen = my_age >= 18
is_sixteen = my_age >= 16

print(is_eighteen)
print(is_sixteen)

if is_sixteen and is_eighteen:
    print("Waiter: Sure, I'll bring you one.")
elif is_sixteen:
    print("Waiter: Sorry, you're too young to drink schnaps, but can I offer you a beer?") 
else:
    print("Waiter: Sorry, you're too young kid!")
````

### ... but there are also wrong ways..


````python
my_age = 18
print("Me: A schnaps, please!")

is_eighteen = my_age >= 18
is_sixteen = my_age >= 16

print(is_eighteen)
print(is_sixteen)

if is_sixteen:
    print("Waiter: Sorry, you're too young to drink schnaps, but can I offer you a beer?") 
elif is_eighteen:
    print("Waiter: Sure, I'll bring you one.")
else:
    print("Waiter: Sorry, you're too young kid!")
````

#### Order of conditions matters

Even though the code *runs*, it might not behave as you expect â€” because **Python checks conditions in order**.

- Python checks is_sixteen first.
- But since is_sixteen is also True when youâ€™re 18 (because 18 â‰¥ 16), the first condition already matches â€” so the second one is never reached.

> Python does exactly what you tell it to do â€” not what you mean.

> Tip: Put the stricter condition first. That way, the more specific case (age â‰¥ 18) is handled before the broader one (age â‰¥ 16).

## For loops

Use a **for loop** when you know how many items you have (or youâ€™re iterating over a collection).

- Over items (values): `for drink in drinks:`
- Over numbers: `for i in range(5):` â†’ `0..4`
- Need index + value: `for i, drink in enumerate(drinks):`

Understanding `for` loops
- A **`for` loop** tells Python to **repeat** a block of code for each item in a collection.
- Instead of writing the same line multiple times, you let Python *do it for you*.
- Think of a for loop like reading names off a list â€” one at a time until the list is finished.

### For each element in a list

````python
drinks = ["beer", "water", "schnaps", "cola"]

for drink in drinks:
    print(drink)
````

### For each element in a range

````python
for i in range (5):
    print(i)
````

````python
for i in range(0, 10): # from 0 until 10 (exclusive)
    print(i)
````

````python
for i in range (2, 6):
    print(i)
````

### Dinner example

````python
print("Put 1 plate on the table.")
print("Put 2 plate on the table.")
print("Put 3 plate on the table.")
print("Put 4 plate on the table.")
````

````python
for n in range(1, 5):
    print(f"Put {n} plate on the table.")
````

#### Bonus example for motivated students

````python
numbers = [1, 2, 3, 4]
sum_numbers = 0
for a_number in numbers:
    sum_numbers = sum_numbers + a_number
    
print(sum_numbers)
````

**Summing numbers with a `for` loop.**

Step-by-step:
1. Start with `sum_numbers = 0`.
2. The loop picks the first number (`1`) â†’ adds it to `sum_numbers` â†’ now `1`.
3. Next number (`2`) â†’ `sum_numbers` becomes `3`.
4. Then `3` â†’ total is `6`.
5. Finally `4` â†’ total is `10`.

After the loop finishes, we print the final total.

**Concepts:**
- Loops can **accumulate** values.
- `sum_numbers` acts as a **counter** or **accumulator** that stores partial results.
- Each loop iteration updates the variable, so it *remembers* the previous result.


### Command: continue

`continue` skips the rest of the current loop iteration and moves on to the next one.

> Use it when you want to *ignore* some cases but keep looping

````python
drinks_to_order = ["beer", "water", "schnaps", "cola"]

for drink in drinks_to_order:
    
    if drink == "water":
        continue
    
    print(drink)
````

### Bonus example for motivated students (if statement within a loop)

````python
drinks_to_order = ["beer", "water", "schnaps", "cola"]
my_age = 16

print("Me: I would like to order the following drinks...")

for drink in drinks_to_order:
    print(f"Me: One {drink}, please.")
    
    if drink == "beer" and my_age < 16:
        print(f"Waiter: Sorry, you are too young to drink beer.")
    elif drink == "schnaps" and my_age < 18:
        print(f"Waiter: Sorry, you are too young to drink schnaps.")
    else:
        print(f"Waiter: Ok, I'll bring you a {drink}.")
````

#### Combining loops and conditions

Here, we use a **loop** to go through each drink and an **if/elif/else** structure to decide how the waiter should respond.

**What happens:**
- The loop repeats for every drink in the list.
- Inside the loop, the `if` statements check **different conditions** depending on the drink and age.
- Each iteration can follow a *different path* based on those conditions.

## While loops

Use **while** when you don't know in advance how many steps are needed â†’ when the stopping point depends on a condition that changes over time.

- Always ensure something inside the loop changes, otherwise you get an infinite loop.
- Think in terms of *state* (e.g., `money`) and *update rule*.

> Tip: print state each iteration to see progress.

````python
money = 9
beer_count = 0
beer_price = 3

while money >= beer_price:
    print(f"Got {money}â‚¬ and drank {beer_count} beers.")
    
    money = money - beer_price
    beer_count = beer_count + 1
````

````python
money = 9
beer_count = 0
beer_price = 3

while money >= beer_price:
    money = money - beer_price
    beer_count = beer_count + 1
    
    print(f"Got {money}â‚¬ and drank {beer_count} beers.")
    
# what is the first output?
# what is the last output?
````


In a loop, **where** you place `print()` decides **when** the information is shown.

- In the **first example**, `print()` comes **before** updating `money` and `beer_count`,  
  so it shows the *state before drinking* each beer.

- In the **second example**, `print()` comes **after** updating the values,  
  so it shows the *state after drinking* each beer.
  
> Each loop iteration runs top to bottom â€” so the order of statements directly affects what output you see.


### Command: break

`break` exits the nearest loop immediately. Useful to stop once a condition is met.

> Pair with a clear message so you can see *why* the loop stopped.

> Safety: add a counter and a `break` in early experiments.

````python
money = 50
beer_count = 0
beer_price = 3

while money >= beer_price:
    print(f"Got {money}â‚¬ and drank {beer_count} beers.")
    
    if beer_count == 5:
        print("ðŸ¤®")
        break
    
    money = money - beer_price
    beer_count = beer_count + 1
````

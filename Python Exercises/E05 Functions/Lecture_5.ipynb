# Classes and Objects

In Python (and many other languages), we often work with **data** and **actions**.

- Data â†’ information (like a carâ€™s color, number of seats, horsepower)  
- Actions â†’ what we can do with it (like drive, honk, stop)

When programs get bigger, itâ€™s useful to group these together â€”  
and thatâ€™s where **Object-Oriented Programming (OOP)** comes in.

---

**Object-Oriented Programming (OOP)** is a way of organizing code where you model **real-world things** as **objects** in your program.

Each object:
- **Has data** (called *attributes*)  
- **Can do things** (called *methods*)

For example:
> A `Car` has attributes like `horse_power` and `n_seats`, and methods like `turn_engine_on()` or `blow_horn()`.

---

### Class vs. Object

- A **class** is like a **blueprint** â€” it defines what a type of object looks like and what it can do.  
  *(like a car design in a factory)*  

- An **object** (or **instance**) is one **actual example** created from that blueprint.  
  *(like a specific car made from that design â€” your car!)*  

You can make many cars from one blueprint â€” each with different values (color, power, etc.), but all following the same structure.

````python
class Car:
    def __init__(self, number_seats, horse_power):
        self.horse_power = horse_power
        self.n_seats = number_seats
        
    def turn_engine_on(self):
        print(f"Motor sound of {self.horse_power} horse power.")
        
    def blow_horn(self):
        print("ðŸŽº")
        
    def print_self(self):
        print(self)
````

### Whatâ€™s happening here:
- `class Car:` â†’ defines a new **type** of object.
- `def __init__(...)` â†’ a **constructor** â€” automatically runs when you create a new `Car`.
- `self` â†’ a reference to *this particular object*.
- `self.horse_power` â†’ stores data inside the object.
- Functions inside a class are called **methods**.

> ðŸ’¡ Think of a class as a *cookie cutter*, and objects as the *cookies* made from it.

## Creating and using an object

````python
my_car = Car(number_seats = 5, horse_power = 400)
````

Now `my_car` is an **instance** (object) of the `Car` class â€” it has its own `horse_power` and `n_seats`.

You can access data (called **attributes**) and behavior (called **methods**) using dot `.` notation:

````python
print(my_car.n_seats)
````

````python
my_car.turn_engine_on()
````

Thatâ€™s why you define methods as `def turn_engine_on(self):`  
even though you call them without `self`:

````python
my_car.blow_horn()
````

````python
my_car.print_self()
````

````python
print(my_car)
````

````python
my_second_car = Car(number_seats = 3, horse_power = 500)
````

## Printing the object

When you print an object directly:

Youâ€™ll see something like:
```
<__main__.Car object at 0x000001F5...>
```

Thatâ€™s Python showing you the *memory location* of the object,  
not a human-friendly description.

To customize this, define a special method `__str__` inside the class:

````python
class Car:
    def __init__(self, number_seats, horse_power):
        self.horse_power = horse_power
        self.n_seats = number_seats
        
    def turn_engine_on(self):
        print(f"Motor sound of {self.horse_power} horse power.")
        
    def blow_horn(self):
        print("ðŸŽº")
        
    def print_self(self):
        print(self)

    def __str__(self):
        return f"Car with {self.n_seats} seats and {self.horse_power} HP"
````

````python
my_car = Car(number_seats = 5, horse_power = 400)
````

````python
print(my_car)
````

````python
my_second_car = Car(number_seats = 3, horse_power = 500)
````

````python
print(my_second_car)
````

### When to use classes

**Use classes when:**
- You have data + behavior that logically belong together
(e.g., a Car has both attributes like horse_power and actions like turn_engine_on()).
- You will create multiple similar objects (e.g., car1, car2, car3).
- Your program is getting large and needs organization and structure.
- You want to reuse or extend functionality later (inheritance, specialized versions).

**Donâ€™t use classes when:**
- Your script is simple and procedural (few functions and variables).
- You donâ€™t need to create multiple instances of something.

> Rule of thumb: If youâ€™re describing â€œthingsâ€ with data and behavior, use a class.

# Inheritance: building on existing classes

So far, each class stood on its own.  
But what if you want a new class that is *just like another*, only with some extra features?

Thatâ€™s where **inheritance** comes in.

**Inheritance** means one class can *reuse* and *extend* another class.  
The new class is called a **child** (or **subclass**),  
and the one it builds upon is the **parent** (or **base class**).

Itâ€™s like saying:

> A **Cylinder** *is also a* **Circle**, but with one extra feature â€” a **height** and a method to calculate **volume**.

````python
class Circle():
    def __init__(self, r):
        self.r = r
        
    def get_radius(self):
        return self.r

    def get_area(self):
        return 3.14 * self.r**2
````

This Circle class can calculate its radius and area.

Now we create a Cylinder that inherits from Circle:

````python
class Cylinder(Circle):
    def __init__(self, r, h):
        Circle.__init__(self, r)
        self.h = h

    def get_volume(self):
        return self.get_area() * self.h
````

The Cylinder class inherits all methods from Circle,
so it already knows how to calculate area.

Then it adds one new thing: height (h) and a method to calculate volume.

````python
small_cylinder = Cylinder(r = 10, h = 10)
print(small_cylinder.get_area())
print(small_cylinder.get_volume())
````

- `get_area()` comes from the parent class (Circle)
- `get_volume()` is new to the child class (Cylinder)

````python
big_circle = Circle(r = 100)
print(big_circle.get_area())
````

### Whatâ€™s happening under the hood:

When you write:

```python
class Cylinder(Circle):
```

you tell Python:

> Start with everything from Circle, and then add or override as needed.

So:

- You reuse the parentâ€™s methods (e.g., get_area()).
- You can extend it (add get_volume()).
- You can also override existing methods to change behavior (not done here, but possible).

Why inheritance is useful:

- Avoid repetition: Donâ€™t rewrite code that already works.
- Logical hierarchy: A cylinder is a circle with extra dimensions.
- Easier maintenance: Fix or improve code in one place (the parent class).
- Extensible design: Add new specialized shapes (like Sphere) later easily.

You can also use the built-in super() function instead of explicitly calling the parent class:

````python
class Cylinder(Circle):
    def __init__(self, r, h):
        super().__init__(r)
        self.h = h

    def get_volume(self):
        return self.get_area() * self.h
````

````python
small_cylinder = Cylinder(r = 10, h = 10)
print(small_cylinder.get_area())
print(small_cylinder.get_volume())
````

### Example: Car â†’ ElectricCar

Letâ€™s use **Car** class and extend it.

````python
class Car:
    def __init__(self, brand, fuel_type):
        self.brand = brand
        self.fuel_type = fuel_type

    def start_engine(self):
        print(f"{self.brand} engine starts â€” {self.fuel_type} power!")


class ElectricCar(Car):
    def __init__(self, brand, battery_capacity):
        super().__init__(brand, fuel_type="electric")
        self.battery_capacity = battery_capacity

    def charge(self):
        print(f"{self.brand} is charging the battery ({self.battery_capacity} kWh).")
````

````python
my_tesla = ElectricCar("Tesla", battery_capacity=75)
my_tesla.start_engine()  # inherited from Car
my_tesla.charge()        # new ElectricCar method
````

- `ElectricCar` *is a* `Car`, but with some extra properties.
- It **reuses** all logic from `Car` using `super()`.
- This avoids duplicating code like engine startup behavior.

### Example: Animal â†’ Dog â†’ Puppy

This example shows **multi-level inheritance**,  
where one class inherits from another, which itself inherits from another.

````python
class Animal:
    def eat(self):
        print("Eating...")

class Dog(Animal):
    def bark(self):
        print("Woof!")

class Puppy(Dog):
    def play(self):
        print("Playing fetch!")
````

````python
p = Puppy()
p.eat()   # from Animal
p.bark()  # from Dog
p.play()  # from Puppy
````

- `Puppy` inherits from `Dog`, which inherits from `Animal`.  
- So `Puppy` automatically gets **all methods** from both parent classes.  
- This shows that inheritance can form a â€œchainâ€ â€” each level adds something new.

# Working with Exceptions

When something goes wrong in your program, Python **raises an exception** â€”  a special type of error that interrupts normal code execution.

An **exception** is Pythonâ€™s way of saying:
> â€œSomething unexpected happened â€” I canâ€™t continue unless you handle this!â€

````python
e = Exception("My custom error")
print(e.args)
````

`"My custom error"` is stored as a tuple in `e.args`.

You can pass multiple values if you like:

````python
e = Exception("My custom error", 1, 2)
print(e.args)
````

To trigger (or raise) an exception manually, use the `raise` keyword:

````python
raise Exception("My custom error")
````

````python
raise Exception("test")
````

This immediately stops the program and displays your error message.


You raise custom exceptions when:
- You want to signal a specific problem in your code logic.
- You want to stop execution early and make the error message clear.

### Catching exceptions

So far, we saw what happens when an error is *raised* â€” the program **stops** and prints a traceback.  
But what if you want your code to **keep running**, even if something goes wrong?

Thatâ€™s where **`try` / `except`** comes in.

You â€œtryâ€ to run some code, and if it fails, Python jumps to the **`except`** block.

````python
open("/tmp/non_existing_file", "r")
print("we still work")
````

````python
try:
    open("/tmp/non_existing_file", "r")
except FileNotFoundError as e:
    print(f"Caught: {e}")
    
print("we still work")
````

- Code inside `try` runs normally.
- If it raises an exception of type `FileNotFoundError`, the program jumps to except.
- The variable `e` contains the exception object â€” with the error message inside.

The program continues after the `except` block â€” no crash, no traceback.

Or you can catch any error:

````python
try:
    open("/tmp/non_existing_file", "r")
except Exception as e:
    print(f"Something went wrong: {e}")
````

You can handle different errors separately.

````python
try:
    open("/tmp/non_existing_file", "r")
except ValueError as e:
    print(f"Caught a ValueError: {e}")
except FileNotFoundError as e:
    print(f"Caught a FileNotFoundError: {e}")

print("we still work")
````

### Try - Except - Else

You can also use an **`else`** block â€” it runs *only if no error happens*.

````python
try:
    print("all good")
except NameError: # this runs if that specific error happens
    print("Undefined variables found.")
except: # this runs if *any* other error happens
    print("Don't know this error.")
else: # this runs only if no error occurred
    print("Everything is fine!")
````

Why use `else`?

Using `else` helps separate:
- Code that might fail (in `try`)
- Code that runs only if it didnâ€™t fail (in `else`)

It makes your programs more readable and avoids confusion.

**Produce a NameError exception**

````python
try:
    print(a) # Undefined variable
    print("all good")
except NameError:
    print("Undefined variables found.") # catch and continue
except:
    print("Don't know this error.")
else:
    print("Everything is fine!")
    
print("normal program flow")
````

**Produce an FileNotFoundError exception**

The generic except catches the unknown error, prints a message, then raises it again (so itâ€™s visible in the traceback).

````python
try:
    print("all good")
    open("/tmp/non_existing_file", "r")
except NameError:
    print("Undefined variables found.")
except:
    print("Don't know this error.")
    raise # throw the exception "upwards"
else:
    print("Everything is fine!")
    
print("normal program flow")
````

### Finally

Runs no matter what.

````python
try:
    print("a")
except FileNotFoundError:
    print("File does not exist")
except:
    print("Don't know this error")
    #raise
finally:
    print("cleaning up")
````

### Type checks

Sometimes, you want to make sure that a you work with **the correct type of input**.  
For example, your code might only make sense for *integers*, not strings or floats.

````python
def check_integer(number):
    if isinstance(number, int):
        print(f"My number: {number}")
    else:
        raise TypeError("Only integers are allowed as input.")
````

If the input is not an `int`, Python raises a custom `TypeError` with a helpful message.
This is better than a generic crash â€” it explains what the user did wrong.

````python
check_integer(2)
````

````python
check_integer(2.0)
````

````python
def check_numeric(number):
    if isinstance(number, (int, float)):
        print(f"My number: {number}")
    else:
        raise TypeError("Only integers / floats are allowed as input.")
````

````python
check_numeric(2.0)
````

````python
check_numeric(2)
````

````python
check_numeric("2")
````

```python
# Write your code in this cell:
def calculate_area_of_circle(d: float = 30) -> float:
    """calculates the area of a circle

    Args:
        d (float, optional): diameter

    Returns:
        float: Area of the circle (rounded to two decimal places)
    """    
    a = round(math.pi*(d/2)**2,2)
    return a
```

```python
# Write your code in this cell:
def calculate_grade(is_a1_submitted: bool, 
                    is_a2_submitted: bool, 
                    a1_points: float, 
                    a2_points: float, 
                    lp_points: float, 
                    exam_points: float, 
                    bonus_points: float) -> int | None:
    """calculates the grade of a student

    Args:
        is_a1_submitted (bool): True if Assignment 1 was submitted, False otherwise.
        is_a2_submitted (bool): True if Assignment 2 was submitted, False otherwise.
        a1_points (float): The points received for Assignment 1.
        a2_points (float): The points received for Assignment 2.
        lp_points (float): The points received for the Live Programming.
        exam_points (float): The points received for the lecture exam.
        bonus_points (float): The total amount of bonus points received.

    Returns:
        int | None: The calculated grade if a grade was received, 
        None if the student was deregistered from the course without a grade.
    """     
    if is_a1_submitted is False:
        print("​​​​As you didn't submit Assignment 1, you were deregistered from the course and receive no grade.")
        return None
    elif is_a2_submitted is False:
        print("​​​​As you didn't submit Assignment 2, your Informatics 1 grade is 5.")
        return int(5)
    elif exam_points <= 7.5:
        print("​​​​As you didn't pass the Lecture Exam, your Informatics 1 grade is 5.")
        return int(5)
    #elif is_a1_submitted is True and is_a2_submitted is True and exam_points >
    else:
        practical_part_points = a1_points + a2_points + lp_points
        total_points = practical_part_points + exam_points
        if total_points > 50 and practical_part_points <= 80:
            practical_part_points += bonus_points
            total_points = practical_part_points + exam_points
        if total_points > 90:
            grade = int(1)
            print(f"​​​​You have {total_points} total points, and your Informatics 1 grade is {grade}.")
            return grade
        elif total_points > 80:
            grade = int(2)
            print(f"​​​​You have {total_points} total points, and your Informatics 1 grade is {grade}.")
            return grade
        elif total_points > 65:
            grade = int(3)
            print(f"​​​​You have {total_points} total points, and your Informatics 1 grade is {grade}.")
            return grade
        elif total_points > 50:
            grade = int(4)
            print(f"​​​​You have {total_points} total points, and your Informatics 1 grade is {grade}.")
            return grade
        else:
            grade = int(5)
            print(f"​​​​You have {total_points} total points, and your Informatics 1 grade is {grade}.")
            return grade
```

### 3.1 Wine barrel analysis

```python
# Write your code in this cell:
def check_wine_barrels(barrels: list[float]) -> str:
    """Prints the emptiest barrel as well as the fullest barrel and returns the average fill level

    Args:
        barrels (list[float]): The list of fill levels.

    Returns:
        str: The average fill level across all barrels rounded to 2 decimal places.
    """    
    # for value in barrels:
    #     min_value = barrels[0]
    #     if value < min_value:
    #         min_value = value
    min_value = min(barrels)
    min_position = barrels.index(min_value) + 1
    max_value = max(barrels)
    max_position = barrels.index(max_value) + 1
    sum_value = sum(barrels)
    number_list = len(barrels)
    avg_value = str(round(sum_value / number_list, 2))
    print(f"​​​​The barrel on position {min_position} was the emptiest with {min_value} l.")
    print(f"​​​​The barrel on position {max_position} was the fullest with {max_value} l.")
    return avg_value + " l"
```

### 3.2 Tailoring 
#### Required Function: `dressmaking_orders`

```python
# Write your code in this cell:
def dressmaking_orders(items: list[str],
                       prices: list[float],
                       status: list[str]) -> list[list[str | float]]:
    """Does the tailoring of the shop.

    Args:
        items (list, optional): The items of clothing
        prices (list, optional): The prices of the items of clothing
        status (list, optional): The sales status of the items of clothing

    Returns:
        list: The new list with the partially modified prices
    """    
    new_list = []
    sold_count = 0
    for it, pri, stat in zip(items, prices, status):
        if stat == "sold":
            sold_count += 1
            continue
        elif stat == "in stock":
            new_price = round(pri * 0.9, 2)
            new_list.append([it, new_price, stat])
        elif stat == "reserved":
            new_list.append([it, pri, stat])

    print (f"{sold_count} items of clothing were sold.")
    return new_list
```

### 3.3 Ski trips
#### Required Function: `ski_compare`

```python
# Write your code in this cell:
def ski_compare(austria: set[str], switzerland: set[str]) -> tuple[set[str], set[str], set[str], set[str]]:
    """Answers different questions about ski trips either to Austria of Switzerland

    Args:
        austria (set[str]): The list of participants for the trip to Austria
        switzerland (set[str]): The list of participants for the trip to Switzerland

    Returns:
        tuple[set[str], set[str], set[str], set[str]]: 
        The answers to the four questions described above as tuples of sets, in the correct order.
    """    
    only_austria = austria - switzerland
    one_destination = austria ^ switzerland
    both = austria & switzerland
    at_least_one = austria | switzerland
    return only_austria, one_destination, both, at_least_one
```

## 4. Dictionaries
### 4.1 Hiking trails
#### Required Function: `find_trails`

```python
# Write your code in this cell:
def find_trails(trails: dict[str, dict[str, int]], 
                max_length: int, 
                max_difficulty: int) -> tuple[dict[str, int], dict[str, int]]:
    """Filter function for trails according max_lenght and max_difficulty

    Args:
        trails (dict[str, dict[str, int]]): 
        A dictionary containing the names of the hiking trails as keys and their information as values.
        max_length (int): A maximum length (in kilometres) that should not be exceeded.
        max_difficulty (int): A maximum difficulty level (1–5) that should not be exceeded.

    Returns:
        tuple[dict[str, int], dict[str, int]]: A tuple with 2 dictionaries as described above
    """    
    matching_trails = {trail : info["length"] 
                       for trail, info in trails.items()
                       if info["length"] <= max_length and info["difficulty"] <= max_difficulty
                       }
    none_matching_trails = {trail : info["length"]
                            for trail, info in trails.items()
                            if info["length"] > max_length or info["difficulty"] > max_difficulty

    }
    if matching_trails:
        for trail, length in matching_trails.items():
            print(f"{trail}: {length} km, Difficulty {trails[trail]['difficulty']}")
    else:
        print(f"No trails with a length of at most {max_length} km and a difficulty of at most {max_difficulty} found.")
    return (matching_trails, none_matching_trails)
```

```python
# Test your code in this cell:
trails = {
    'Alpine trail': {'length': 12, 'difficulty': 3},
    'Lake circuit': {'length': 8, 'difficulty': 2},
    'Drachenwand via ferrata': {'length': 5, 'difficulty': 5},
    'Forest path': {'length': 3, 'difficulty': 1},
    'Summit route': {'length': 14, 'difficulty': 4}
}
max_length = 10
max_difficulty = 2

find_trails(trails, max_length, max_difficulty)
```

## 5.1 Strings
### 5.1 Tutor's Secret Message
#### Required Function: `decode_tutor_message`

```python
# Write your code in this cell:
def decode_tutor_message(encoded_lines: list[str], code_words: dict[str,str], noise_symbols: tuple[str]) -> str:
    """The functions decodes the message from the tutor.

    Args:
        encoded_lines (list[str]): The encrypted message.
        code_words (dict[str,str]): Mapping from abbreviations or codewords to full words.
        noise_symbols (tuple[str]): Characters to remove completely from the text.

    Returns:
        str: The decoded and cleaned message from the tutor.
    """    
    
    encoded_lines_wo_noise = []
    for line in encoded_lines:
        for char in noise_symbols:
            line = line.replace(char, "")
        encoded_lines_wo_noise.append(line)
        
    decoded_lines = []
    for line in encoded_lines_wo_noise:
        words = line.split()
        new_words = [code_words.get(word, word) for word in words]
        decoded_lines.append(" ".join(new_words))

    message = ". ".join(decoded_lines) + "."

    sentences = [s.strip().capitalize() for s in message.split(".") if s.strip()]
    message = ". ".join(sentences) + "."

    return message
```

# Example 1

## Type casting

```python
number = int("123")
pi = float("3.14159")
int_pi = int(pi)
ready = bool(12)
pi_str = str(pi)

print(f"{number} of type {type(number)}")
print(f"{pi} of type {type(pi)}")
print(f"{int_pi} of type {type(int_pi)}")
print(f"{ready} of type {type(ready)}")
print(f"{pi_str} of type {type(pi_str)}")
```

## Lists revisited

```python
drinks = ["beer", "schnaps", "cola"]
print(drinks)
drinks.append("pepsico")
print(drinks)
```

```python
# add elements to the end of the list using append()
drinks.append("water")
```

```python
print(drinks)
```

```python
# remove the first matching element of the list using remove()
drinks.remove("schnaps")
```

```python
print(drinks)
```

```python
drinks = ["beer", "schnaps", "cola", "schnaps"]
print(drinks)
# check if it really only removes the first occuring element
drinks.remove("schnaps")
print(drinks) 
```

```python
# remove the last element using pop()
drinks.pop()
print(drinks)
```

```python
drinks = ["beer", "cola", "schnaps"]
drinks.pop(0) # drops the first element (0-based index)
print(drinks)
```

```python
drinks.pop(1) # drops the second element, what is the output?
print(drinks)
```

```python
drinks = ["beer", "cola", "schnaps"]
n_drinks = len(drinks) # get length of the list
print(n_drinks)
```

```python
drinks = ["beer", "cola", "schnaps"]
print(drinks.pop(0))
print(drinks)
```

## Tuples

```python
print(stages_of_life[0])
print(stages_of_life[1])
print(stages_of_life[-1]) # negative indexing
print(stages_of_life[-2])
```

```python
stages_of_life.append("hell") # what happens if we want to add an element?
```

## Sets

```python
my_interests = {"gambling", "music", "drinking"}
friends_interests = {"sports", "music", "cooking"}
print(my_interests.intersection(friends_interests))
print(my_interests.union(friends_interests))
```

```python
my_interests = ["gambling", "music", "drinking", "gambling"]
my_interests_set = set(my_interests) # cast the list to a set for removing duplicates

print(my_interests)
print(my_interests_set)
```

# Example 5

## Functions

```python
def say_hello_function():
    print("Hello from a function")
```

```python
say_hello_function()
```

```python
def say_hello_to_friend(friends_name):
    print(f"Hello {friends_name}")
```

```python
say_hello_to_friend("Peter")
```

```python
say_hello_to_friend("Frank")
```

```python
def introduce_myself(friends_name, my_name="Tom"):
    print(f"Hello {friends_name}, I'm {my_name}.")
```

```python
introduce_myself("Peter")
introduce_myself("Peter", "Niklas") # positional arguments
introduce_myself(friends_name="Peter", my_name="Niklas") # keyword arguments
introduce_myself(my_name="Niklas", friends_name="Peter") # order does not matter with keyword arguments
```

```python
introduce_myself("Niklas", "Peter") # What is the output?
```

```python
# define function with arbitrary parameters
def greet_all_friends(*friends):
    for friend in friends:
        print(f"Hi {friend}!")
        
greet_all_friends("Peter", "Thomas")
```

# Example 6

## Return values

```python
def pythagoras(a, b):
    return a**2 + b**2
```

```python
c_squared = pythagoras(2, 5)
print(c_squared)
```

```python
def pythagoras_with_input(a, b):
    c_squared = a**2 + b**2
    return (a, b, c_squared)
```

```python
result = pythagoras_with_input(5, 5)
print(result)
```

```python
def pythagoras_print(a, b):
    print(a**2 + b**2)
```

```python
pythagoras_print(2, 5)
```

```python
print(f"The result of 2^2 + 5^2 = {pythagoras(2, 5)}")
```

```python
print(f"The result of 2^2 + 5^2 = {pythagoras_print(2, 5)}")
```

# Example 7

## Variable scope

```python
a = 0          # This is a global variable 
if a == 0: 
    b = 1      # This is still a global variable 

def my_function(c): 
    d = 3      # this is a local variable 
    print(f"c={c} d={d}") 

my_function(7) # Call the function, passing 7 as parameter 

print(f"a={a} b={b}") # a and b still exist 
print(c)       # c doesn't exist anymore --> error
print(d)       # d doesn't exist anymore --> error
```

### Watch out where you define variables!

```python
e = 1          # This is a global variable 
if e == 0: 
    f = 1      # This is also a global variable

print(f"e={e} f={f}") # what is the output?
```
